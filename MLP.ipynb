```python
import numpy as np
import pandas as pd
```

```python
import tensorflow as tf
```

```python
import keras
```

# Load Data

```python
data=np.load('./processed_data/since_2010_processed_data.npy')
```

```python
results=np.load('./processed_data/since_2010_results.npy')
```

```python
inds=list(range(len(data)))
```

```python
np.random.shuffle(inds)
```

```python
X_train=data[inds[:int(len(inds)*.8)]]
X_test=data[inds[int(len(inds)*.8):]]
y_train=results[inds[:int(len(inds)*.8)]]
y_test=results[inds[int(len(inds)*.8):]]
```

# MLP Modeling

```python
from keras.models import Sequential
from keras.layers import Dense, Activation,Dropout
from keras.callbacks import ModelCheckpoint
```

```python
model=Sequential()
model.add(Dense(32,input_dim=100,activation='relu'))
model.add(Dropout(.2))
model.add(Dense(32,activation='relu'))
model.add(Dropout(.2))
model.add(Dense(8,activation='relu'))
model.add(Dropout(.2))
model.add(Dense(1,activation='sigmoid'))
```

```python
import keras.backend as K

def mean_pred(y_true, y_pred):
    return K.mean(y_pred)
```

```python
sgd = keras.optimizers.SGD(lr=0.01, decay=1e-6, momentum=0.9, nesterov=True)
```

```python
model.compile(optimizer=sgd,loss='binary_crossentropy',metrics=['accuracy',mean_pred])
```

```python
checkpointer = ModelCheckpoint(filepath='./models/8_1.h5',monitor='val_loss', verbose=1, save_best_only=True)
```

```python
model.fit(X_train,y_train,validation_data=[X_test,y_test],callbacks=[checkpointer],batch_size=32,epochs=500)
```
